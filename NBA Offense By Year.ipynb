```python
import pandas as pd

# read CSV file into a Pandas DataFrame
df = pd.read_csv(r'C:\Users\tyler\Downloads\NBA_Team_Stats.csv')

# group the data by year and team
year_team_pts = df.groupby(['Year', 'Team']).sum()['Pts']

# create an empty dataframe to store the team with the most points and the average points for each year
max_pts_teams = pd.DataFrame(columns=['Team', 'Avg Pts', 'Diff'])

# find the team with the most points and the average points for each year
for year in df['Year'].unique():
    year_team_pts = df[df['Year'] == year].groupby('Team')['Pts'].sum()
    team_with_max_pts = year_team_pts.idxmax()
    max_pts = year_team_pts.loc[team_with_max_pts]
    avg_pts = df[df['Year'] == year]['Pts'].mean()
    diff = max_pts - avg_pts
    max_pts_teams.loc[year] = [team_with_max_pts, avg_pts, diff]

# sort the results by the difference between a team's points and the league average
max_pts_teams = max_pts_teams.sort_values(by='Diff', ascending=False)

# display the result with the team name, average points, and difference
for year, team in max_pts_teams.iterrows():
    team_with_max_pts = team['Team']
    year_team_pts = df[df['Year'] == year].groupby('Team')['Pts'].sum()
    max_pts = year_team_pts[team_with_max_pts]
    avg_pts = team['Avg Pts']
    diff = team['Diff']
    print(f"In {year}, {team_with_max_pts} averaged {max_pts:.1f} points. The league average was {avg_pts:.1f}. The difference is {diff:.1f} points.")
```

```python

import pandas as pd
import matplotlib.pyplot as plt

# read CSV file into a Pandas DataFrame
df = pd.read_csv(r'C:\Users\tyler\Downloads\NBA_Team_Stats.csv')

# group the data by year and team
year_team_pts = df.groupby(['Year', 'Team']).sum()['Pts']

# create an empty dataframe to store the team with the most points and the average points for each year
max_pts_teams = pd.DataFrame(columns=['Team', 'Avg Pts', 'Diff'])

# find the team with the most points and the average points for each year
for year in df['Year'].unique():
    year_team_pts = df[df['Year'] == year].groupby('Team')['Pts'].sum()
    team_with_max_pts = year_team_pts.idxmax()
    max_pts = year_team_pts.loc[team_with_max_pts]
    avg_pts = df[df['Year'] == year]['Pts'].mean()
    diff = max_pts - avg_pts
    max_pts_teams.loc[year] = [team_with_max_pts, avg_pts, diff]

# create the bar plot
fig, ax = plt.subplots()
ax.bar(max_pts_teams.index, max_pts_teams['Diff'], label='Difference')


# add labels and a legend to the graph
ax.set_xlabel('Year')
ax.set_ylabel('Difference')
ax.set_title('NBA Team Points Difference')
ax.legend()

# set the x-axis ticks and labels to show only the last part of the year
ax.set_xlabel('Year')
ax.set_xticklabels([str(year)[-2:] for year in max_pts_teams.index])

# display the graph
plt.show()




# create the line plot
fig, ax = plt.subplots()
ax.plot(max_pts_teams.index, max_pts_teams['Diff'], label='Difference')

# shade the area below the line
ax.fill_between(max_pts_teams.index, max_pts_teams['Diff'], color='lightblue')

# add labels and a legend to the graph
ax.set_xlabel('Year')
ax.set_ylabel('Difference')
ax.set_title('NBA Team Points Difference')
ax.legend()

# set the x-axis ticks and labels to show only the last part of the year
ax.set_xlabel('Year')
ax.set_xticklabels([str(year)[-2:] for year in max_pts_teams.index])

# display the graph
plt.show()


# create the scatter plot
fig, ax = plt.subplots()
ax.scatter(max_pts_teams.index, max_pts_teams['Diff'], label='Difference')

# add labels and a legend to the graph
ax.set_xlabel('Year')
ax.set_ylabel('Difference')
ax.set_title('NBA Team Points Difference')
ax.legend()

# set the x-axis ticks and labels to show only the last part of the year
ax.set_xlabel('Year')
ax.set_xticklabels([str(year)[-2:] for year in max_pts_teams.index])

# display the graph
plt.show()


# create the line plot
fig, ax = plt.subplots()
ax.plot(max_pts_teams.index, max_pts_teams['Diff'], marker='o', label='Difference')


# add labels and a legend to the graph
ax.set_xlabel('Year')
ax.set_ylabel('Difference')
ax.set_title('NBA Team Points Difference')
ax.legend()

# set the x-axis ticks and labels to show only the last part of the year
ax.set_xlabel('Year')
ax.set_xticklabels([str(year)[-2:] for year in max_pts_teams.index])

# display the graph
plt.show()


# create the stacked bar plot
fig, ax = plt.subplots()
ax.bar(max_pts_teams.index, max_pts_teams['Avg Pts'], label='Average Points')
ax.bar(max_pts_teams.index, max_pts_teams['Diff'], bottom=max_pts_teams['Avg Pts'], label='Difference')

# add labels and a legend to the graph
ax.set_xlabel('Year')
ax.set_ylabel('Points')
ax.set_title('NBA Team Points')
ax.legend()

# set the x-axis ticks and labels to show only the last part of the year
ax.set_xlabel('Year')
ax.set_xticklabels([str(year)[-2:] for year in max_pts_teams.index])

# display the graph
plt.show()
```

```python
# Group the data by player and year
groups = NBA_Offense_with_Season.groupby(['PPG_Leader', 'Year'])

# Filter the groups where player was both Pts_Leader and PPG_Leader
scoring_groups = [(group[0], group[1]) for group in groups 
             if (group[1]['Pts_Leader'].iloc[0] == group[1]['PPG_Leader'].iloc[0])]

# Sort the filtered groups by year (most recent to earliest)
sorted_groups = sorted(scoring_groups, key=lambda x: x[0][1], reverse=True)

# Print the PPG_Leader, PPG_Leader_Team, Year, and PPG_Season for each group
for group in sorted_groups:
    ppg_leader = group[0][0]
    ppg_leader_team = group[1]['PPG_Leader_Team'].iloc[0]
    year = group[0][1]
    ppg_season = group[1]['PPG_Season'].iloc[0]
    display(f"Year: {year}, {ppg_leader}, {ppg_leader_team}, PPG: {ppg_season}")
```

```python
from collections import Counter

# Find the unique years where each PPG_Leader appeared
unique_years_ppg_leader = NBA_Offense_Champions.loc[NBA_Offense_Champions['PPG_Leader'].notnull(),  ['Year', 'PPG_Leader']].groupby('Year')['PPG_Leader'].apply(set)

# Find the unique years where each PPG_Player appeared
unique_years_ppg_lead = NBA_Offense_Champions.loc[NBA_Offense_Champions['PPG_Player'].notnull(),  ['Year', 'PPG_Player']].groupby('Year')['PPG_Player'].apply(set)

# Find the unique years where each Finals MVP appeared
unique_years_finals_mvp = NBA_Offense_Champions['Year'].loc[NBA_Offense_Champions['Finals_MVP'].notnull()].unique()
unique_years_finals_mvp_players = NBA_Offense_Champions.loc[NBA_Offense_Champions['Finals_MVP'].notnull(), ['Year', 'Finals_MVP']].groupby('Year')['Finals_MVP'].apply(set)

# Combine the sets of players for each category
all_ppg_leaders = set.union(*unique_years_ppg_leader)
all_ppg_players = set.union(*unique_years_ppg_lead)
all_finals_mvp_players = set.union(*unique_years_finals_mvp_players)

# Count the occurrences of each player and assign a value to the count
ppg_leader_counts = Counter(player for year in unique_years_ppg_leader for player in year)
ppg_leader_counts = {player: 5*count for player, count in ppg_leader_counts.items()}

ppg_player_counts = Counter(player for year in unique_years_ppg_lead for player in year)
ppg_player_counts = {player: 3*count for player, count in ppg_player_counts.items()}

finals_mvp_counts = Counter(player for year in unique_years_finals_mvp_players for player in year)
finals_mvp_counts = {player: 2*count for player, count in finals_mvp_counts.items()}

# Combine the dictionaries and total the unique counts for each player
all_counts = {player: ppg_leader_counts.get(player, 0) + ppg_player_counts.get(player, 0) + finals_mvp_counts.get(player, 0) for player in set.union(all_ppg_leaders, all_ppg_players, all_finals_mvp_players)}

# Sort the dictionary by value in descending order
sorted_counts = sorted(all_counts.items(), key=lambda x: x[1], reverse=True)

# Print the results
print("Total unique counts for all players:")
for player, count in sorted_counts:
    print(f"{player}: {count}")

```
